# Составляем рейтинг моделей

Посчитав наши метрики, мы высчитываем среднее по каждой метрики и потом считаем среднее всех метрик. 

```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/karpovapolina/Downloads/gen_poems_metrics_fin.csv')
```

```python
df
```

```python
new_df = df.drop('text', axis=1)
```

```python
avg_scores = new_df.groupby('model').mean()

rankings = avg_scores.rank(ascending=False)

avg_rankings = rankings.mean(axis=1)

sorted_models = avg_rankings.sort_values(ascending=False)

print("Models ranked by performance:")
print(sorted_models)
```
